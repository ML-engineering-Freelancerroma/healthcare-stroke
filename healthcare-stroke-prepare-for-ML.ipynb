```python
import pandas as pd

df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
df['gender'].value_counts()
```

```python
import numpy as np


n_rows = len(df)
n_to_change = int(n_rows * 0.3)

indices_to_change = np.random.choice(df.index, size=n_to_change, replace=False)

lowercase_map = {
    'Male': 'male',
    'Female': 'female',
}

df.loc[indices_to_change, 'gender'] = df.loc[indices_to_change, 'gender'].map(lowercase_map)

df['gender'].value_counts()
```

```python
df['stroke'].value_counts()
```

```python
from sklearn.model_selection import train_test_split


df_remaining, df_sample_25 = train_test_split(
    df,
    test_size=0.25,
    stratify=df['stroke'],
    random_state=42
)

df_remaining = df_remaining.reset_index(drop=True)
df_sample_25 = df_sample_25.reset_index(drop=True)

df_sample_25.to_csv('healthcare-stroke-data-test.csv', index=False)
df_remaining.to_csv('healthcare-stroke-data-train.csv', index=False)
```
